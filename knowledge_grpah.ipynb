```python
# pip install sparqlwrapper
# https://rdflib.github.io/sparqlwrapper/

import sys
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint_url = "https://query.wikidata.org/sparql"

query = """PREFIX wd: <http://www.wikidata.org/entity/>
PREFIX schema: <http://schema.org/>

SELECT ?item ?itemLabel ?desc
WHERE
{
  ?item wdt:P289 wd:Q674909.
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}"""


def get_results(endpoint_url, query):
    user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
    # TODO adjust user agent; see https://w.wiki/CX6
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()


results = get_results(endpoint_url, query)

for result in results["results"]["bindings"]:
    print(result)
```

URN: https://en.wikipedia.org/wiki/Uniform_Resource_Name 

```python
wiki_data = {
    "DDG-76": "https://www.wikidata.org/wiki/Q1474743",
    "DDG-96": "https://www.wikidata.org/wiki/Q2301237",
    "DDG-115": "https://www.wikidata.org/wiki/Q7873256",
    "DDG-69":"https://www.wikidata.org/wiki/Q1669467",
    "DDG-75":"https://www.wikidata.org/wiki/Q201238",
    "CVN-78":"CVN-78",
    "DDG-77":"DDG-77",
    "DDG-65":"DDG-65",
    "DDG-71":"DDG-71",
    "DDG-84":"DDG-84",
    "DDG-58":"DDG-58"
}

for index, row in testing_df.iterrows():
    failure_id = row["id"]
    # bob = URIRef(u'urn:bob')
    node = rdflib.URIRef("urn:uuid:{}".format(failure_id))
    kg.add(node, kg.get_ns("rdf").type, kg.get_ns("base").Failure_Event) 

    # Agent and Agent Role nodes
    agent_role_node = rdflib.URIRef("urn:kg:{}".format(row["sfrSubmittedBy"]))
    # agent_role_node = rdflib.URIRef("https://w3id.org/la3d/ont/failureevent#Agent_Role/{}".format(row["sfrSubmittedBy"]))
    kg.add(node, kg.get_ns("base").provides_agent_role, agent_role_node)
    kg.add(agent_role_node, kg.get_ns("rdf").type, kg.get_ns("base").Agent_Role)

    agent_node = rdflib.URIRef("urn:kg:{}".format(row["sfrSubmittedBy"]))
    kg.add(agent_role_node, kg.get_ns("base").performed_by, agent_node)
    kg.add(agent_node, kg.get_ns("rdf").type, kg.get_ns("base").Agent)

    # Temporal entity
    # temporal_entity_node = rdflib.URIRef("urn:kg:{}".format(row["fiscalYear"]))
    # kg.add(node, kg.get_ns("base").At_Time, temporal_entity_node)
    # kg
    # interval_node = rdflib.URIRef("urn:kg:{}".format(row["fiscalYear"]))
    # kg.add(temporal_entity_node, kg.get_ns("base").Interval, interval_node)

    # Place and information object Nodes
    location_node = rdflib.URIRef(rdflib.URIRef("urn:kg:{}".format(row["siteName"])))
    kg.add(node, kg.get_ns("base").has_location, location_node)
    kg.add(location_node, kg.get_ns("rdf").type, kg.get_ns("base").Place)

    information_object_node = rdflib.URIRef(rdflib.URIRef("urn:kg:{}".format(row["siteName"])))
    kg.add(location_node,kg.get_ns("base").is_described_by,information_object_node)
    kg.add(information_object_node, kg.get_ns("rdf").type, kg.get_ns("base").Information_Object)

    primary_indentifer_node = rdflib.URIRef(rdflib.URIRef("urn:kg:{}".format(row["siteName"])))
    kg.add(information_object_node, kg.get_ns("base").has_primary_identifier, primary_indentifer_node)
    kg.add(primary_indentifer_node, kg.get_ns("rdf").type, kg.get_ns("base").Identifier)

    kg.add(information_object_node, kg.get_ns("base").has_description, rdflib.Literal(row["siteName"]))
    kg.add(information_object_node, kg.get_ns("base").seealso, rdflib.Literal(wiki_data[row["siteName"]]))


```
